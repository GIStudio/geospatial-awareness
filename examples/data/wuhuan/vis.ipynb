```python
import geopandas as gpd
import matplotlib.pyplot as plt

def read_and_plot_shp(file_path):
    # SHP
    data = gpd.read_file(file_path)
    
    # 
    data.plot()
    plt.title('Shapefile Visualization')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.show()

shapefile_path = "./District_boundary.shp"  # SHP
read_and_plot_shp(shapefile_path)

shapefile_path = "/workspace/programming-language-demo/geospatial-awareness/examples/data/wuhuan/road.shp"  # SHP
read_and_plot_shp(shapefile_path)
```
